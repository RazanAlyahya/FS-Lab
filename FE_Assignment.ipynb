```python
import pandas as pd 
import numpy as np
import category_encoders as ce

```

```python
data=pd.DataFrame({'height':['Low','Medium','High','Low','Medium','High','Low','Medium','High'],
                  'name':  ['Razan','Atheer','Norah','Mohammed','Razan','Atheer','Mohammed','layla','Mohammed'],
                  'Month': ['January','April','March','April','Februay','June','July','June','September']
                 })
```

# Q1: Apply Ordinal encoding.

## global variable

```python
from sklearn.preprocessing import OrdinalEncoder
#ord1 = OrdinalEncoder()

encoder= ce.OrdinalEncoder(cols=['height'],return_df=True,mapping=[{'col':'height','mapping':{'None':0,'High':1,'Medium':2,'Low':3}}])

# transforming the column after fitting
data["height"] = encoder.fit_transform(data[["height"]])

data.head(10)
```

## local variable

# Q2: Apply One hot encoding.

## global variable

```python
from sklearn.preprocessing import OneHotEncoder

# define one hot encoding
encoder= OneHotEncoder()

encoder = encoder.fit_transform(data[['name']]).toarray()
# converting arrays to a dataframe
encoded_colm = pd.DataFrame(encoder)

# concatenating dataframes
data = pd.concat([data, encoded_colm], axis=1)

# removing the encoded column.
#data = data.drop(['name'], axis=1)

data.head(10)
```

## local variable

```python
from sklearn.preprocessing import OneHotEncoder
# define data
dataOneHot=pd.DataFrame({'name':[
'Razan','Atheer','Norah','Mohammed','Razan','Atheer','Mohammed'
]})

# define one hot encoding
encoder = OneHotEncoder(sparse=False)
# transform data
onehot = encoder.fit_transform(dataOneHot)
print(onehot)
```

# Q3: Apply Frequency Encoding. 

## global variable

```python
# grouping by frequency
df = data.groupby('name').size()/len(df)

# mapping values to dataframe
data.loc[:, "name_freq_encode"] = data['name'].map(df)

# drop original column.
#data = data.drop(['nom_0'], axis=1)

data.head(10)
```

## local variable

```python
from sklearn.preprocessing import LabelEncoder 

dataFrequency=pd.DataFrame({'name':[
'Razan','Atheer','Norah','Mohammed','Razan','Atheer','Mohammed'
]})

le = LabelEncoder()
dataFrequency['Frequency'] = le.fit_transform(dataFrequency['name'])


print(dataFrequency)
```
